```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import mutual_info_regression
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate dummy data
n_samples = 1000

# Feature 1: linear relationship
X1 = np.linspace(0, 10, n_samples)

# Feature 2: sinusoidal relationship
X2 = np.sin(X1)

# Feature 3: random noise
X3 = np.random.rand(n_samples)

# Construct target variable with some noise
y = 2*X1 + 0.5*np.sin(3*X2) + np.random.normal(0, 0.5, n_samples)

# Combine features into a DataFrame
data = pd.DataFrame({'X1_linear': X1, 'X2_sine': X2, 'X3_random': X3})
```

```python
data.plot();
```

```python
mi = mutual_info_regression(data, y)
mi_series = pd.Series(mi, index=data.columns)
```

```python
mi_series.sort_values(ascending=False).plot(kind='bar')
plt.ylabel('Mutual Information')
plt.show()
```
